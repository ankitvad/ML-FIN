```python
NDX = pd.io.data.get_data_yahoo('^NDX', 
                                 start=datetime.datetime(1994, 12, 1), 
                                 end=datetime.datetime(2014, 12, 1))

FTSE = pd.io.data.get_data_yahoo('^FTSE', 
                                 start=datetime.datetime(1994, 12, 1), 
                                 end=datetime.datetime(2014, 12, 1))

GSPC = pd.io.data.get_data_yahoo('^GSPC', 
                                 start=datetime.datetime(1994, 12, 1), 
                                 end=datetime.datetime(2014, 12, 1))
```

Load 20 years of daily prices of FTSE 100, S&P 500 and NASDAQ-100. 
They are plotted below.
(Green: FTSE 100; Blue: NASDAQ-100; Red: S&P 500)

```python
NDX_1d_prices = NDX['Adj Close']
NDX_1d_prices.plot()

FTSE_1d_prices = FTSE['Adj Close']
FTSE_1d_prices.plot()

GSPC_1d_prices = GSPC['Adj Close']
GSPC_1d_prices.plot()
```

Shift time index by -1. 

```python
import statsmodels.api as sm

NDX_1d_prices_shifted = NDX_1d_prices.shift().dropna()
FTSE_1d_prices_shifted = FTSE_1d_prices.shift().dropna()
GSPC_1d_prices_shifted = GSPC_1d_prices.shift().dropna()
```

Here are plotted autocorrelation functions for all three time series in order to determine if these time series are weakly dependent.

For how many lags? (for 100 currently, looks like dropping linearly) => KUBO HELP ME, HELP ME, I AM LOST WITHOUT YOU :D

(colors same as before)

```python
plt.plot(sm.tsa.stattools.acf(NDX_1d_prices, nlags=100))
plt.plot(sm.tsa.stattools.acf(FTSE_1d_prices, nlags=100))
plt.plot(sm.tsa.stattools.acf(GSPC_1d_prices, nlags=100))
plt.show()
```

Estimate AR(1) processes for all three stock indexes. Results are in the tables below.

```python
y = NDX_1d_prices[1:]
X = NDX_1d_prices.shift().dropna()
X = sm.add_constant(X) 

print 'for NASDAQ:'

est = sm.OLS(y, X)
est = est.fit()
est.summary()
```

```python
y = FTSE_1d_prices[1:]
X = FTSE_1d_prices.shift().dropna()
X = sm.add_constant(X) 

print 'for FTSE 100:'
est = sm.OLS(y, X)
est = est.fit()
est.summary()
```

```python
y = GSPC_1d_prices[1:]
X = GSPC_1d_prices.shift().dropna()
X = sm.add_constant(X) 

print 'for S&P 500:'

est = sm.OLS(y, X)
est = est.fit()
est.summary()
```

```python
import statsmodels.api as sm
```

```python
y = NDX_1d_returns[1:5035]
X = NDX_1d_returns.shift().dropna()
X = sm.add_constant(X)  # Adds a constant term to the predictor
X.head()
```
