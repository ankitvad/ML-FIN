```python
ret_Gold_m = GOLD.pct_change(periods=1).dropna()['1951-01-01':]
ret_Gold_1Q = GOLD.pct_change(periods=3).dropna()['1951-01-01':]
ret_Gold_2Q = GOLD.pct_change(periods=6).dropna()['1951-01-01':]
ret_Gold_4Q= GOLD.pct_change(periods=12).dropna()['1951-01-01':]
print len(ret_Gold_4Q) == len(ret_Gold_2Q) == len(ret_Gold_1Q) == len(ret_Gold_m)
print len(ret_Gold_4Q)
```

Ako sa meni STDEV pri zmene casoveho ramcu: 
(0: monthly, 1:1Q, etc.)

http://en.wikipedia.org/wiki/Jules_Regnault
"the deviation of prices is directly proportional to the square root of time".

```python
STDEV_ALL = [ret_Gold_m.std(), ret_Gold_1Q.std(),ret_Gold_2Q.std(),ret_Gold_4Q.std()]
plt.plot(STDEV_ALL)
plt.xlabel('Month to Year')
plt.ylabel('Standard Deviation')
```

```python
MEAN_ALL = [ret_Gold_m.mean(), ret_Gold_1Q.mean(),ret_Gold_2Q.mean(),ret_Gold_4Q.mean()]
plt.plot(MEAN_ALL)
plt.xlabel('Month to Year')
plt.ylabel('Mean')
```
