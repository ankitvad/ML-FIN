## Random Walk of Gold

```python
GOLD = pd.read_csv('GOLD.csv', index_col='date', parse_dates=True)
GOLD.tail()
```

```python
GOLD.plot()
```

Does Gold price follow random walk?

*   http://en.wikipedia.org/wiki/Random_walk_hypothesis
*   http://en.wikipedia.org/wiki/Random_walk

```python
GOLD_x = GOLD[:-1] 
GOLD_y = GOLD[1:]
```

```python
import statsmodels.api as sm

X = np.array(GOLD_x)
y = np.array(GOLD_y)

est = sm.OLS(y, X)
est = est.fit()
est.summary()
```

Signifakntny koeficient => Random Walk Confirmed
